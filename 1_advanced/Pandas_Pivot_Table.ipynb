```python
# Importing Libraries
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

Count of Jobs


```python
df.pivot_table(index='job_title_short',  aggfunc='size')
```

```python
df.groupby('job_title_short').size()
```

MEDIAN SALARY BY JOB

```python
df.pivot_table(values='salary_year_avg',index='job_title_short', aggfunc='median').sort_values(by='salary_year_avg', ascending=False)
```

```python
df.groupby('job_title_short')['salary_year_avg'].median().sort_values(ascending=False)
```

```python
df.pivot_table(values = 'salary_year_avg', index = 'job_country',columns = 'job_title_short' , aggfunc = 'median')
```

```python
top_countries = df['job_country'].value_counts().head(10).index

df_job_country_salary = df.pivot_table(
    values='salary_year_avg',
    index='job_country',
    columns='job_title_short',
    aggfunc='median'        
)

df_job_country_salary = df_job_country_salary.loc[top_countries]

job_titles = ['Data Scientist', 'Data Engineer', 'Data Analyst']

df_job_country_salary = df_job_country_salary[job_titles]

df_job_country_salary.plot(kind='bar')
plt.ylabel('Median Salary (USD)')
plt.xlabel('')
plt.title('Median Salary by Job Country and Job Title')
plt.xticks(rotation=45, ha='right')
plt.show()
```
