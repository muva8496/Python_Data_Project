```python
# Importing Libraries
import ast
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

#Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['job_skills'] = df['job_skills'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x)
```

```python
df['job_work_from_home'].value_counts().plot(kind='pie', startangle=90, autopct='%1.3f%%')
plt.title("Work From Home")
plt.ylabel('')
plt.show()
```

```python
df['job_title_short'].value_counts().plot(kind='pie')
plt.title("Hob_titles")
plt.ylabel('')
plt.show()
```

```python
fig, ax = plt.subplots()

ax.pie(df['job_work_from_home'].value_counts(), labels=df['job_work_from_home'].value_counts().index, autopct='%1.3f%%', startangle=90)
```

```python
lfig, ax = plt.subplots(1, 3, figsize=(15, 5))  # ✅ subplots not subposts

dict_column = {
    'job_work_from_home': 'Work From Home',
    'job_no_degree_mention': 'No Degree Mention',
    'job_health_insurance': 'Health Insurance'
}

for i, (column, title) in enumerate(dict_column.items()):  # ✅ complete for loop
    ax[i].pie(                                              # ✅ ax not as
        df[column].value_counts(),                          # ✅ loop variable not hardcoded
        labels=['True', 'False'],
        autopct='%1.1f%%'
    )
    ax[i].set_title(title)                                 # ✅ add title to each

plt.tight_layout()
plt.show()
```
