```python
# Importing Libraries
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

#Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
df_ke = df[df['job_country'] == 'Kenya'].copy()
```

```python
df_ke
```

```python
df_ke['job_posted_month'] = df['job_posted_date'].dt.strftime('%B')

df_ke
```

```python

df_ke_pivot = df_ke.pivot_table(index='job_posted_month', columns='job_title_short', aggfunc='size')

df_ke_pivot
```

```python
df_ke_pivot.reset_index(inplace = True)
df_ke_pivot['month_no']= pd.to_datetime(df_ke_pivot['job_posted_month'], format='%B').dt.month

df_ke_pivot['month_no']
```

```python
df_ke_pivot['month_no'] = pd.to_datetime(
    df_ke_pivot['job_posted_month'], format='%B'
).dt.month

df_ke_pivot.sort_values('month_no', inplace=True)
df_ke_pivot.set_index('job_posted_month', inplace=True)
df_ke_pivot.drop(columns='month_no', inplace=True)
```

```python
df_ke_pivot
```

```python
df_ke_software_pivot =  pd.read_csv("https://lukeb.co/software_csv", index_col = 'job_posted_month')

df_ke_software_pivot
```

```python
df_ke_merged = df_ke_pivot.merge(df_ke_software_pivot, on = 'job_posted_month')
```

```python
top_5 = (df_ke_merged.
         sum()
         .sort_values(ascending=False)
         .head(5)
         .index
         .sort_values())

top_5
```

```python
df_ke_merged[top_5].plot(kind='line')
plt.title('Top 5 Job Titles in Kenya')
plt.xlabel('Month')
plt.ylabel('Jobs_count')
plt.ylim(0, 20000)
plt.legend()
plt.show()
```
