```python
# Importing Libraries
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()  

# Data Cleaning
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
df_ke = df[df['job_country'] == 'Kenya'].copy()
```

```python
df_ke
```

```python
df_ke['job_posted_month'] = df['job_posted_date'].dt.strftime('%B')

df_ke
```

```python

df_ke_pivot = df_ke.pivot_table(index='job_posted_month', columns='job_title_short', aggfunc='size')

df_ke_pivot
```

```python
df_ke_pivot.reset_index(inplace = True)
df_ke_pivot['month_no']= pd.to_datetime(df_ke_pivot['job_posted_month'], format='%B').dt.month

df_ke_pivot['month_no']
```

```python
df_ke_pivot['month_no'] = pd.to_datetime(df_ke_pivot['job_posted_month'], format='%B').dt.month
df_ke_pivot.sort_values('month_no', inplace=True)
df_ke_pivot.set_index('job_posted_month', inplace=True)
df_ke_pivot.drop(columns='month_no', inplace=True)
```

```python
df_ke_pivot
```

```python
df_ke_pivot.plot(kind='line')
```

```python
top_3 = df_ke['job_title_short'].value_counts().head(3)
top_3.index.tolist()

top_3
```

```python
df_ke_pivot[top_3.index].plot(kind='line')
plt.title('Trends of Top 3 Job Titles in Kenya Over Months')
plt.xlabel('2023')
plt.ylabel('Job Count')
plt.legend()
plt.show()
```
