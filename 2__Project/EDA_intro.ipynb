ROLES TO EXPLORE


```python
# Importing Libraries
import ast
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

#Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['job_skills'] = df['job_skills'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x)
```

```python
df_DE_Canada = df[(df['job_country'] == 'Canada') & (df['job_title_short'] == 'Data Analyst')]
```

JOB LOCATIONS

```python
import seaborn as sns
```

```python
df_plot = df_DE_Canada['job_location'].value_counts().head(10).to_frame()

sns.set_theme(style='ticks')
sns.barplot(data=df_plot, x='count',y='job_location', hue='count', palette='dark:b_r', legend=False)
sns.despine()
plt.title('Number of Jobs per Location')
plt.xlabel('Count of Jobs')
plt.ylabel('')
plt.show()
```

JOB OPPORTUNITIES

```python
lfig, ax = plt.subplots(1, 3, figsize=(15, 5))  # ✅ subplots not subposts

dict_column = {
    'job_work_from_home': 'Work From Home',
    'job_no_degree_mention': 'No Degree Mention',
    'job_health_insurance': 'Health Insurance'
}

for i, (column, title) in enumerate(dict_column.items()):  # ✅ complete for loop
    ax[i].pie(                                              # ✅ ax not as
        df_DE_Canada[column].value_counts(),                          # ✅ loop variable not hardcoded
        labels=['True', 'False'],
        autopct='%1.1f%%'
    )
    ax[i].set_title(title)                                 # ✅ add title to each

plt.tight_layout()
plt.show()
```

COMPANIES TO EXPLORE


```python
df_plot = df_DE_Canada['company_name'].value_counts().head(10).to_frame()

sns.set_theme(style='ticks')
sns.barplot(data=df_plot, x='count',y='company_name', hue='count', palette='dark:b_r', legend=False)
sns.despine()
plt.title('Count of Companies for DE in Canada')
plt.xlabel('Count of Jobs')
plt.ylabel('')
plt.show()
```
